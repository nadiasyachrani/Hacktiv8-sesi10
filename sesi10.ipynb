```python
# manual
n = 86+226
pro_fm = 226/n
print(n,pro_fm)
```

```python
#std error dari female
stderror_fm = (pro_fm*(1-pro_fm)/n)**(1/2)
stderror_fm
```

```python
# confident error 95% (z score 1.96)
z_score = 1.96
low_ci_fm = pro_fm - (z_score*stderror_fm)
up_ci_fm = pro_fm + (z_score*stderror_fm)
print(low_ci_fm,up_ci_fm)
```

### visualisasikan area dari confident interval



```python
stats.norm.cdf(2.17)
```

```python
z = 1.96
```

```python
lower_bound = mean_of_sample_means - z * std_error
lower_bound
```

```python
upper_bound = mean_of_sample_means + z * std_error
upper_bound
```

```python
kde = stats.gaussian_kde(sample_means) #gaussian=normal
pos = np.linspace(np.min(sample_means),np.max(sample_means),10000)
plt.plot(pos,kde(pos))
shade = np.linspace(lower_bound,upper_bound,300)
plt.fill_between(shade,kde(shade),alpha=0.5)
plt.xlabel('sample mass (pounds)')
plt.ylabel('frequency')
plt.axvline(x=mean_of_sample_means,color='red')
plt.text(x=154,y=.01,s='shaded area are \npart of the confidence \nintervals 95%')
```

## HYPOTHESIS TESTING IN POPULATION MEANS
* menguji hipotesis apakah var a memiliki hubungan satistik var b
* melihat variabel yang berpengaruh, contoh kolesterol sama sakit jantung ada hubungannya?

```python
df = pd.read_csv('nhanes_2015_2016.csv')
df
```

#### uji apakah rata2 BMI dari laki2 lebih tinggi dari BMI perempuan

h0  = rata2 BMI laki2 tidak lebih tinggi signifikan dari perempuan

h1 = rata2 BMI laki2 lebih tinggi signifikan dari perempuan

----
h0 : meanL <= meanP

H1 : meanL > meanP

RIAGENDR female = 2 Male =1

```python
females = df[df.RIAGENDR==2]
males = df[df.RIAGENDR==1]
```

```python
# jumlah data, mean, stdev
n_females = len(females)
mean_females = females['BMXBMI'].mean()
std_females = females['BMXBMI'].std()
print(n_females,mean_females,std_females)
```

```python
# jumlah data, mean, stdev
n_males = len(males)
mean_males = males['BMXBMI'].mean()
std_males = males['BMXBMI'].std()
print(n_males,mean_males,std_males)
```

```python
females.BMXBMI.isna().sum()
```

h0 : meanL <= meanP

H1 : meanL > meanP

karena confidence level 95% maka alpha = 0.05

```python
import statsmodels.api as sm
```

```python
alpha = 0.05
bmi_hypo = sm.stats.ztest(males['BMXBMI'].dropna(),females['BMXBMI'].dropna(),alternative='larger')
bmi_hypo
```

```python
bmi_hypo[1]
```

```python
if bmi_hypo[1]>alpha:
    print('accept nul hypotesis (ho) mean of males BMI is not larger than female')
else:
    print('reject null hypotesis (h0)  mean of males BMI is larger than female')
```

#### data sama hipotesis berbeda

UJI APAKAH MEAN BMI LAKI2 TIDAK SAMA DENGAN PEREMPUAN

H0 : meanL = meanP

H1 : meanL != meanP

variabel bebas gender, terikatnya laki2

```python
bmi_hypo2 = sm.stats.ztest(males['BMXBMI'].dropna(),females['BMXBMI'].dropna(),alternative='two-sided')
bmi_hypo2
```

```python
if bmi_hypo2[1]>alpha:
    print('accept nul hypotesis (ho)')
else:
    print('reject null hypotesis (h0)')
```

## HYPOTHESIS TESTING IN POPULATION PROPORTION


```python
n = 1018
pro_null = 0.52
pro_alt = 0.56
```

```python
pro_hypo = sm.stats.proportions_ztest(n*pro_alt,n,pro_null,alternative='larger')
pro_hypo
```

```python
if pro_hypo[1]>alpha:
    print('accept nul hypotesis (ho)')
else:
    print('reject null hypotesis (h0)')
```
